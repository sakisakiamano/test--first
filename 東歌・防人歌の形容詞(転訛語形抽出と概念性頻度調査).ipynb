**東歌の形容詞(概念性頻度調査) byテキスト処理**

---
---

##### 操作手順

---
- 語彙素読みを重複ありでリスト化
  - この際、一緒に取り出してしまったインデクスは除いておく
---
- リストのデータを使って頻度表作成(ディクショナリ機能)
- 項目と頻度を並べて出力




```
from google.colab import files
uploaded = files.upload()
```

```
#形容詞の概念性(カナシの多さの証明)
#語彙素読み列をまずは重複ありでリストに入れる
columnlist = []
inputfile = open('東歌の形容詞.txt')
for line in inputfile:
  columns = line.split(',')
  columnlist.append(columns[1])

#インデクスの'語彙素読み'を用例に入れないようにする
columnlist.remove('語彙素読み')

#語彙素読みリストから頻度表作成
freq = {}
for goiso in columnlist:
  if goiso in freq:
    freq[goiso] += 1
  else:
    freq[goiso] =1

for word in sorted(freq, key=freq.get, reverse=True):
  print(word + '\t' + str(freq[word]))
```
